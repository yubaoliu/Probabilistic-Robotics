```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import expon
from scipy import stats
```

```python
expect_stuck_time=60
stuck_pdf = expon(scale=expect_stuck_time)
```

```python
zs = np.arange(0, 100, 0.1)
ys = [stuck_pdf.pdf(z) for z in zs]

plt.plot(zs,ys)
plt.title("probaility density function")
plt.show()
```

```python
expected_escape_time=60
escape_pdf = expon(scale=expected_escape_time)
```

```python
zs2 = np.arange(0, 100, 0.1)
ys2 = [escape_pdf.pdf(z) for z in zs2]

plt.plot(zs2,ys2)
plt.title("probaility density function")
plt.show()
```
