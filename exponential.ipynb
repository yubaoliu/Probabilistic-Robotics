```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import expon
from scipy import stats
```

```python
stone_number = 1.0
expected_travel_distance = 1/stone_number
p = expon(scale=expected_travel_distance)
zs = np.arange(0, 5, 0.01)
ys = [p.pdf(z) for z in zs]

plt.plot(zs,ys)
plt.title("probaility density function")
plt.show()
```

```python
stone_number = 2.0
expected_travel_distance = 1/stone_number
p = expon(scale=expected_travel_distance)
zs = np.arange(0, 5, 0.01)
ys = [p.pdf(z) for z in zs]

plt.plot(zs,ys)
plt.title("probaility density function")
plt.xlabel("Distnance")
plt.ylabel("pdf")
plt.show()
```

```python
p = stats.expon.cdf(0.5, scale=1/1)
print(p)
```
